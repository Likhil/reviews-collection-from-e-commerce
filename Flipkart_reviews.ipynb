```python
#importing required packages.
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from pprint import pprint
from bs4 import BeautifulSoup
#starting the chrome driver
driver = webdriver.Chrome()
#opening the homepage
driver.get('https://www.flipkart.com/')
#sleep for 10 secs to load the page fully
time.sleep(10)
#finding the search bar through xpath
search_field = driver.find_element_by_xpath("""//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input""")
print(search_field.text)
#sending keys to search bar
search_field.send_keys('smartphone' + '\n')
#sleep to load results
time.sleep(10)
#closing login window popup
try:
    login_pop = driver.find_element_by_xpath("""/html/body/div[2]/div/div/button""")
    # Here .click function use to tap on desire elements of webpage
    login_pop.click()
    print('pop-up closed')
except:
    pass
#initializing a dictionary to store the links
store={}
time.sleep(10)
#extracting the page info.
page_html = driver.page_source
#converting to html parser
soup1 = BeautifulSoup(page_html, 'html.parser')
soup1.prettify()
#finding the a tags where name price rating and link are expected
for i in soup1.findAll('a',href=True, attrs={'class':'_31qSD5'}):
    name=i.find('div', attrs={'class':'_3wU53n'})
    price=i.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
    rating=i.find('div', attrs={'class':'hGSR34'})
    link=i['href']
    if(rating==None):
        continue
    store[name.text]=[link,price.text,rating.text]
pg_cnt=2
soup_r=soup1
#count is set to 217 as the number of pages
while pg_cnt!=217:
    time.sleep(5)
    #finding a tags for name link and rating recursively from page 2
    next_pg=soup_r.find_all('a',href=True,attrs={'class':'_3fVaIS'})
    print(next_pg)
    #Url from the page numbers will be only half so adding home page url to it
    URL = "https://www.flipkart.com"
    n_pg_link=next_pg[-1]['href']
    new_link=URL + n_pg_link
    print(new_link)
    #opening next page
    driver.get(new_link)
    #extracting information from next page
    response=driver.page_source
    #parsing with beautiful soup
    soup_r = BeautifulSoup(response, 'html.parser')  
    for i in soup_r.findAll('a',href=True, attrs={'class':'_31qSD5'}):
        name=i.find('div', attrs={'class':'_3wU53n'})
        price=i.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
        rating=i.find('div', attrs={'class':'hGSR34'})
        link=i['href']
        #ignoring if rating is not present for the phone
        if(rating==None):
            continue
        #storing in a dictionary
        store[name.text]=[link,price.text,rating.text]    
    pg_cnt+=1
    
pprint(len(store))
```

```python
#script which opens each phones page and extract information
#importing packages
from random import randint
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from pprint import pprint
from bs4 import BeautifulSoup
#starting webdriver
driver = webdriver.Chrome()
#initializing final dictionary
final={}
for i,j in store.items():
    #assigning homepage link to a variable to append in next steps
    main_link="https://www.flipkart.com"
    new=main_link+j[0]
    #opening phone's page
    driver.get(new)
    #timer to load
    time.sleep(10)
    #initializing the individual elements to a dictionary
    final.setdefault(i,{})
    
    final[i]['price']=j[1]
    #try except for rating to ignore if not present
    try:
        final[i]['rating']=j[2]
    except:
        pass
    #extracting page information
    page_data = driver.page_source
    #parsing it into html
    soup_ind=BeautifulSoup(page_data, 'html.parser')
    soup_ind.prettify()
    #finding tr tags for specifications
    for k in soup_ind.findAll('tr',href=False, attrs={'class':'_3_6Uyw row'}):
        value=k.find('li',attrs={'class':'_3YhLQA'}).text
        par=k.find('td',attrs={'class':'_3-wDH3 col col-3-12'}).text
        final[i][par]=value
    name=soup_ind.find('span',attrs={'class':'_35KyD6'}).text
    final[i]['name']=name
    print(name)
    #extracting reviews
    try:
        #finding div tags to open page with reviews
        int1=soup_ind.find('div',attrs={'class':'col _39LH-M'})
        int2=int1.find_all('a',href=True)
        m=int2[-1]['href']
        url_new=main_link+m
        #open reviews page of that product recursively
        driver.get(url_new)
        #initializing the reviews list
        reviews_list=[]
        #extracting first page reviews 
        page_data1 = driver.page_source
        #random timer to mask from blocking
        sleep(randint(10,50))
        soup_ind1=BeautifulSoup(page_data1, 'html.parser')
        #finding and extracting reviews from div tags and appending them to list
        for z in soup_ind1.findAll('div',attrs={'class':'qwjRop'}):
            
            
            reviews_list.append(z.text)
        #extracting second page reviews non recursively as the frst page donot have previos so it will trigger while recursing
        page_2=soup_ind1.find('a',href=True, attrs={'class':'_3fVaIS'})
        page_2_link=page_2['href']
        page_2_total=main_link+page_2_link
        driver.get(page_2_total)
        page2_data=driver.page_source
        sleep(randint(10,25))
        soup_page_2=BeautifulSoup(page2_data, 'html.parser')
        for z in soup_page_2.findAll('div',attrs={'class':'qwjRop'}):
            try:
                driver.find_element_by_xpath("""//*[@id="container"]/div/div[3]/div/div[1]/div[2]/div[5]/div/div/div/div[2]/div/div/span/span""").click()
            except ElementNotVisibleException :
                pass
            reviews_list.append(z.text)
        ite=2
        while ite==2:
            #extracting reviews from pages recursively
            page_ite=soup_page_2.find_all('a',href=True, attrs={'class':'_3fVaIS'})
            if len(page_ite)==2:
                page_ite_link=page_ite[-1]['href']
            else:
                break
            page_ite_total=main_link+page_ite_link
            driver.get(page_ite_total)
            time.sleep(5)
            pageite_data=driver.page_source
            sleep(randint(1,5))
            soup_page_2=BeautifulSoup(pageite_data, 'html.parser')
            ite=len(soup_page_2.find_all('a',href=True, attrs={'class':'_3fVaIS'}))

            for z in soup_page_2.findAll('div',attrs={'class':'qwjRop'}):
                try:
                    driver.find_element_by_xpath("""//*[@id="container"]/div/div[3]/div/div[1]/div[2]/div[5]/div/div/div/div[2]/div/div/span/span""").click()
                except :
                    pass
                reviews_list.append(z.text)

#             if len(reviews_list) >= 50:
#                 break
        final[i]['reviews']=reviews_list
        #writing each record into file independently
        with open('final_data_flipkart.json', 'a') as fp:
            json.dump(final[i], fp)  
    except TypeError:
        continue
    except IndexError:
        continue
```
